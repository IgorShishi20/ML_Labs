# **Apache Pig**

### Краткий обзор:
Pig - это высокоуровневая платформа для создания программ, работающих на Apache Hadoop. При создании приложений с Apache Hadoop, написание всех функций, компилирование и упаковка занимают много времени. Pig повышает уровень абстракции при обработке данных, что значительно ускоряет работу по созданию новых приложений.

### Общий обзор для IT специалиста:
Pig — это платформа для анализа больших наборов данных, состоящая из языка высокого уровня для выражения программ анализа данных в сочетании с инфраструктурой для оценки этих программ. Отличительным свойством программ Pig является то, что их структура допускает существенное распараллеливание, что, в свою очередь, позволяет им обрабатывать очень большие наборы данных.

### Плюсы и минусы:

#### Плюсы:
1. Простота программирования. Добиться параллельного выполнения простых, «поразительно параллельных» задач анализа данных тривиально. Сложные задачи, состоящие из множества взаимосвязанных преобразований данных, явно кодируются как последовательности потоков данных, что упрощает их запись, понимание и обслуживание.
2. Возможности оптимизации. Способ кодирования задач позволяет системе автоматически оптимизировать их выполнение, позволяя пользователю сосредоточиться на семантике, а не на эффективности.
3. Расширяемость. Пользователи могут создавать свои собственные функции для выполнения специальной обработки.

#### Минусы:
1. Ограниченность функционала: Pig Latin является высокоуровневым языком, что может ограничивать возможный функционал. Не всегда желаемый функционал можно уложить в язык Pig Latin.
2. Сложность освоения:  Pig Latin вместе со структурами данных может быть более сложен в освоении, чем другие языки.

### Набор экосистемы для эффективного использования:
1. Pig Latin: Для выполнения запросов к данным на платформе Hadoop.
2. Hadoop Distributed File System (HDFS): Хранение данных для обработки другими инструментами.
3. Apache Kafka: Для работы с потоковыми данными.
4. MapReduce: Для обработки и генерации больших наборов данных с помощью параллельного распределенного алгоритма на кластере.

### Назначение инструмента:
Pig применяется для быстрого создания прототипов распределенных приложений аналитики больших данных с изменчивой структурой таких как веб-логи. Кроме того, Pig может пригодиться в обработке чувствительных ко времени загрузок данных.
