```python
link = urlopen("https://en.wikipedia.org/wiki/Main_Page")
bsh = BeautifulSoup(link.read(),'html.parser')
heading_tags = ["h1", "h2", "h3"]
for tags in bsh.find_all(heading_tags):
  print(tags.name + ' -> ' + tags.text.strip())
```

```python
from random import sample

url = 'https://www.imdb.com/chart/top'

def get_number():
  random_list = sample(range(0, 250), 10)
  return random_list

def get_movie(new_url):
  headers = {
      'User-Agent': 'Mozilla/5.0 (Windows NT 10.0; Win64; x64) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/109.0.0.0 Safari/537.36',
  }
  request = requests.get(new_url,headers = headers)
  bsh = BeautifulSoup(request.text, 'html.parser')
  movies = bsh.find_all(class_="ipc-metadata-list-summary-item")
  return movies

def get_Summary(new_url):
  headers = {
      'User-Agent': 'Mozilla/5.0 (Windows NT 10.0; Win64; x64) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/109.0.0.0 Safari/537.36',
  }
  movie_page = requests.get(new_url,headers = headers)
  soup = BeautifulSoup(movie_page.text, 'html.parser')
  result = soup.find(class_="sc-466bb6c-3 llCpwq").contents[0]
  return result.contents[0]

def get_Movie_Info(movie):
  mov_title = movie.contents[1].contents[0].contents[1].contents[0].contents[0].contents[0].string
  year = movie.contents[1].contents[0].contents[1].contents[1].contents[0].string
  mov_url = 'http://imdb.com'+movie.contents[1].contents[0].contents[1].contents[0].contents[0]['href']
  return mov_title, year, mov_url

def picker():
  ctr = 0
  list = get_number()
  movies = get_movie(url)
  #for i in range(10):
  #  movie = movies[get_number()[i]]
  for movie in movies:
    mov_title, year,mov_url = get_Movie_Info(movie)
    print(mov_title,year,mov_url)
    mov_sum = get_Summary(mov_url)
    print(mov_sum)
    print("----------------")

picker()
```
