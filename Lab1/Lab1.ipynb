```python
import pandas as pd
import matplotlib as ml
import matplotlib.pyplot as plt
```

```python
reader = pd.read_csv('/content/Прошлые данные - USD_RUB (2).csv')
```

```python
import locale
from datetime import datetime

array = reader.to_numpy()
x = array[:,1]
x_converted = []
for i in x:
  i = i.replace(',','.')
  i = float(i)
  x_converted.append(i)
print(x_converted)
y = array[:,0]
y_converted = []
for i in y:
  i = i.replace('.','-')
  i = datetime.strptime(i, '%d-%m-%Y').date()
  y_converted.append(i)
print(y_converted)
```

```python
plt.plot(y_converted,x_converted)
plt.show()
```
